```python
# importing of libaries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
import streamlit as st
```

```python
# setting numeric value to readable format
pd.set_option('display.float_format', '{:,.2f}'.format)
```

```python
df = pd.read_csv("jiji_housing_raw.csv")
df.head(20)
```

```python
# idetifying column that has a missing value
df.isnull().sum()
```

```python
df.dropna(subset= ["region_parent_name"])
```

```python
# cleaning of price column
df["price"] = df["price"].replace(r"[₦,]", "", regex=True)
df["price"]
```

```python
df["price"] = pd.to_numeric(df["price"], errors="coerce")
```

```python
df["price"].astype(float)
```

```python
df['price'].unique()
```

```python
# checking for outliers
Q1 = np.percentile(df["price"], 25)
Q3 = np.percentile(df["price"], 75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

filtered_data = df[(df["price"] >= lower_bound) & (df["price"] <= upper_bound)]
filtered_data
```

```python
# cleaning of region column
df["region"] = df["region"].str.title()
df["region"]
```

```python
# cleaning of region_name column
df["region_name"] = df["region_name"].str.title()
df["region_name"]
```

```python
# cleaning of region_parent_name column
df["region_parent_name"] = df["region_parent_name"].str.title()
df["region_parent_name"]
```

```python
# cleaning of furnishing colmn
df["furnishing"] = df["furnishing"].str.title()
df["furnishing"]
```

```python
# checking for duplicates
df.duplicated().sum()
```

```python
df.to_csv("jiji_housing_cleaned.csv", index=False)
print("Successfully converted Jiji house data.")
```

```python
# the distribution of property sizes
plt.hist(df['property_size'],edgecolor="white", color="skyblue")
plt.xlabel('Property Size')
plt.ylabel('Frequency')
plt.title('Distribution of Property Size')
plt.show()
```

```python
# Do houses with more bedrooms/bathrooms cost significantly more
df.groupby(['bedrooms',"bathrooms"])['price'].mean().sort_values(ascending=False).head(10)
```

```python
# Which regions have the highest number of listings
df["region"].value_counts().sort_values(ascending=False)
```

```python
# Which regions dominate premium property sales?
premium_threshold = df["price"].quantile(0.90)
premium_df = df[df["price"] >= premium_threshold]
premium_counts = premium_df["region"].value_counts()
premium_counts.head(10)
```

```python
# Are furnished apartments more expensive on average
avg = df.groupby('furnishing')['price'].mean()
plt.bar(avg.index, avg.values, color="skyblue", edgecolor="white")
plt.title(" Furnished vs Unfurnished Apartments")
plt.xlabel("Furnishing Status")
plt.ylabel("price")
plt.show()
```

```python

print(df.groupby('is_boost')['price'].mean().sort_values(ascending=False))
```

```python
 # Number of listings per state
state_counts = df['region_parent_name'].value_counts().head(10)
plt.bar(state_counts.index, state_counts.values, color='purple', edgecolor='white')
plt.title("Number of Listings per State")
plt.xlabel("State")
plt.ylabel("Number of Listings")
plt.xticks(rotation=90)
plt.show()
```

```python
 # Average price per state
avg_price = df.groupby('region_parent_name')['price'].mean().sort_values(ascending=False).head(10)
plt.bar(avg_price.index, avg_price.values, edgecolor='white')
plt.title("Average Price per State")
plt.xlabel("State")
plt.ylabel("Average Price")
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of property prices
plt.hist(df['price'])
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Property Prices')
plt.show()

```

```python
 # Price by number of bedrooms
sns.boxplot(x='bedrooms', y='price', data=df)
plt.title('Property Prices by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Property Price (in thousands)')
plt.show()
```

```python
 # Property size vs price
plt.scatter(df['property_size'], df['price'])
plt.title('Property Price vs. Property Size')
plt.xlabel('Property Size (sq. ft / sqm)')
plt.ylabel('Property Price (in thousands)')
plt.show()
```

```python
 # Furnishing type distribution
furnishing_counts = df['furnishing'].value_counts()

plt.pie(furnishing_counts,labels=furnishing_counts.index,autopct='%1.1f%%')
plt.title('Furnishing Type Distribution')
plt.show()
```

```python
# Correlation of numeric variables (price, size, bedrooms, bathrooms
numeric_cols = ["price", "property_size", "bedrooms", "bathrooms"]
corr_matrix = df[numeric_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap="vlag", fmt=".2f", center = 0, linewidth="0.5",linecolor= "blue")
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```
